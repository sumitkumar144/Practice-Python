Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

```python
transactions.join(users,on='UserID',how='left',lsuffix='_left', rsuffix='_right')#Here lsuffix and rsuffix is trying to
#rename the UserID column from left and right table (Without doing that it will throw error). You can also mention either of
# them to make join work
```

```python
import numpy as np
```

```python
transactions.UserID=transactions.UserID.astype('Int64')
```

```python
transactions.UserID=transactions.UserID.astype('Float64')
```

```python
transactions
```

```python
pd.merge(transactions,users,on='UserID',how='left')
```

Which transactions have a UserID not in users?

```python
transactions[~transactions['UserID'].isin(users['UserID'])]
```

Join users to transactions, keeping only rows from transactions and users that match via
UserID (inner join)

```python
pd.merge(transactions,users,how='inner',on='UserID')
```

Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
pd.merge(transactions,users,how='outer',on='UserID')
```

Determine which sessions occurred on the same day each user registered

```python
pd.merge(sessions,users,how='inner',left_on=['UserID','SessionDate'],right_on=['UserID','Registered'])
```

Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
users
```

```python
products
```

```python
users['key']=1
products['key']=1
```

```python
users.merge(products,how='outer',left_on='key',right_on='key')[['UserID','ProductID']]
```

Determine how much quantity of each product was purchased by each user

```python
pd.merge(users,products,how='outer',left_on='key',right_on='key').merge(transactions,how='left',on=['UserID','ProductID']).fillna(value=0)[['UserID','ProductID','Quantity']]
```

For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)

```python
pd.merge(transactions,transactions,on=['UserID'])
```

Join each user to his/her first occuring transaction in the transactions table

```python
data=pd.merge(users,transactions.groupby('UserID').first().reset_index(),how='left',on='UserID')
data
```

Test to see if we can drop columns

```python
test_columns=list(data.columns)
test_columns
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns) #set threshold to drop NAs
```

```python
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing)) #count of missing data
```

```python
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
