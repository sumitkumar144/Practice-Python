#### How do i find and remove duplicate rows in Pandas?

```python
user_cols = ['user_id','age','gender','occupation','zip_code']
users = pd.read_table('http://bit.ly/movieusers',sep='|',header=None,names=user_cols,index_col='user_id')
```

```python
users.head()
```

```python
users.shape
```

Find Out Duplicate Zip Codes

```python
users.zip_code.duplicated()
```

```python
users.zip_code.duplicated().sum() #All the trues and falses added up to give that there are 148 number of duplicates
```

```python
users.duplicated() #It will give a True if there is a row identical to the previous row
```

```python
users.duplicated().sum()
```

```python
users.loc[users.duplicated(),:] # This is to see what are the rows that are duplicate in the dataset
```

```python
users.loc[users.duplicated(keep='first'),:] # Logic for keep="first": Mark Duplicate as True except for the first occurence
```

```python
users.loc[users.duplicated(keep='last'),:] 
```

```python
users.loc[users.duplicated(keep=False),:] 
```

Drop duplicate from the dataFrame

```python
users.drop_duplicates(keep='first').shape
```

```python
users.shape
```

```python
users.drop_duplicates(keep='last').shape
```

```python
users.drop_duplicates(keep=False).shape
```

What if you want to consider a certain columns while identifying duplicates. You don't want to consider all columns in the

```python
users.duplicated(subset=['age','zip_code']).sum()
```

```python
users.loc[users.duplicated(subset=['age','zip_code'])]
```

#### Crime Report Data Set

```python
crime=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
crime.head(3)
```

```python
crime.info()
```

```python
crime.shape
```

```python
crime.dtypes
```

Convert the type of the column Year to datetime64

```python
crime.Year=pd.to_datetime(crime.Year,format="%Y")
```

```python
crime.head(3)
```

```python
crime.dtypes
```

Set the Year column as the index of the dataframe

```python
crime=crime.set_index('Year',drop=True)
crime.head()
```

Group the year by decades and sum the values

```python
crime.resample('10AS').sum()
```

```python
# Uses resample to sum each decade
crimes=crime.resample('10AS').sum()
```

```python
crimes
```

```python
crimes.shape
```

```python
# Uses resample to get the max value only for the "Population" column
population = crime['Population'].resample('10AS').max()
population
```

```python
# Updating the "Population" column
crimes['Population']=population
crimes
```
