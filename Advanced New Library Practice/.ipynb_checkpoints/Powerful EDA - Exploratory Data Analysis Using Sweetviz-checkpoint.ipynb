# Powerful EDA - Exploratory Data Analysis Using Sweetviz

```python
import pandas as pd
```

```python
### Dataset Link : https://www.kaggle.com/c/house-prices-advanced-regression-techniques
import sweetviz
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
train.tail()
```

When there are a lot of unique categorical features, pandas-profiling uses a lot if memory to do the calculation and it might run out of memory. So inthat case sweetviz will help.

```python
%%time
my_report = sweetviz.analyze([train,'Train'], target_feat='SalePrice')
```

```python
my_report.show_html('Report.html')
```

```python
train.describe()
```

#### Comparing Between Two DataFrames(e.g: Test vs Training Sets)

```python
%%time
my_report1 = sweetviz.compare([train,'train'],[test,'test'],'SalePrice')
```

```python
my_report1.show_html('Report1.html')
```
